# Отличия от классической математики:
Python допускает применение сложения и умножения в операциях со строками. Сложение строк называется конкатенацией:

```python
"арбуз" + " -" " это" + " ягода"
```

В Python можно умножать строку на целое число, это называется репликация:

```python
"Арбуз"*4
```

Поскольку память компьютера не бесконечная, поэтому бесконечно повторяющееся представление двоичной дроби 0.1 округляется до конечной дроби. Такие числа называются числа с плавающей точкой.Увидеть значение с плавающей точкой, которое сохраняется для 0.1, можно с помощью метода .as_integer_ratio()

```python
0.2 + 0.1
```
